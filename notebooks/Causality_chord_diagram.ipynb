# Analytics Translator Causality Visualization

```python
in_matrix = pd.read_csv('data/20200106_causality_distance_matrix.csv',sep=';')
in_matrix.head()
```

```python
in_matrix=in_matrix.iloc[19:].set_index('parent')  # cut out the first base layer
in_matrix=in_matrix.iloc[:,4:19]  # cut out all columns base layer
```

```python
in_matrix
```

```python

in_matrix=100-in_matrix*100
in_matrix[in_matrix==50]=0
in_matrix
```

```python
## group and count the index
```

```python
how_many=15 # with base layer 19, without 15
matrix=in_matrix.groupby(level=0).sum()
matrix=matrix.iloc[:,0:how_many]
matrix=matrix.reindex(sorted(matrix.columns), axis=1)
```

```python
matrix
```

```python


#matrix=matrix.astype(int)
d_matrix=np.array(matrix)
np.fill_diagonal(d_matrix, 0)
```

```python
#matrix.columns
```

```python
d_matrix=d_matrix[0:how_many,0:how_many].round(0)
```

```python
#d_matrix[:,0]=1
```

```python
input_chord=d_matrix.tolist()
```

```python
my_labels=list(matrix.columns[0:how_many])
my_labels[11]='Proof of Value' # for wrap labels renaming
```

# chord diagram with standard package

```python
Chord(input_chord, list(matrix.columns[0:how_many]),colors="d3.schemeSet2",
     width=1200,
        label_color="#454545",
        wrap_labels=True,
        margin=140,
        credit=False,
     ).to_html('chord_licence.html')
```

# chord diagram with pro package

```python
Chord(input_chord, my_labels,colors="d3.schemeSet2",
     width=1200,
        label_color="#454545",
        wrap_labels=True, ##wraps the labels when margin is to small
        margin=140,
        credit=True,
        verb="has similarity score of: ",
        noun="",
        font_size_large="30px",
        font_size="24px",
       # title="The Analytics Translator - NLP distance relation between topics",
        arc_numbers=False
     ).to_html('results/Causality_analytics_translator_chord_diagram.html')
```
